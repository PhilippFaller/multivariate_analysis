```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('data/iris.csv', header=None)
data.describe()
```

```python
y = data.values[:, 2]
x = data.values[:, :2]
x
```

```python
import statsmodels.api as sm

x = sm.add_constant(x)
x
```

```python
model = sm.OLS(y, x).fit()
model.summary2()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

b = model.params

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(data[0], data[1], data[2])

x1 = np.arange(data[0].min(), data[0].max(), 0.1)
x2 = np.arange(data[1].min(), data[1].max(), 0.1)
x_grid, y_grid = np.meshgrid(x1, x2)
z_grid = b[0] + b[1]*x_grid + b[2]*y_grid

ax.plot_surface(x_grid, y_grid, z_grid)
```
